```python
import pandas as pd
import matplotlib.pyplot as plt # à¸™à¸³à¹€à¸‚à¹‰à¸²à¹„à¸¥à¸šà¸£à¸²à¸£à¸µà¸ªà¸³à¸«à¸£à¸±à¸šà¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿ

# 1. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¹‚à¸«à¸¥à¸”à¹à¸¥à¸°à¹€à¸•à¸£à¸µà¸¢à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
# *** à¹‚à¸›à¸£à¸”à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸š Path à¸‚à¸­à¸‡à¹„à¸Ÿà¸¥à¹Œ '../data_raw/sample_sales.csv' à¹ƒà¸«à¹‰à¸–à¸¹à¸à¸•à¹‰à¸­à¸‡à¸à¹ˆà¸­à¸™à¸£à¸±à¸™ ***
df = pd.read_csv('../data_raw/sample_sales.csv')

# 2. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¸„à¸³à¸™à¸§à¸“à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'revenue'
df['revenue'] = df['quantity'] * df['unit_price'] * (1 - df['discount'])

# 3. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¹€à¸£à¸µà¸¢à¸‡à¸¥à¸³à¸”à¸±à¸š (à¸­à¸²à¸ˆà¸ˆà¸°à¹„à¸¡à¹ˆà¸ˆà¸³à¹€à¸›à¹‡à¸™à¸ªà¸³à¸«à¸£à¸±à¸šà¸à¸²à¸£à¸žà¸¥à¸­à¸•à¸à¸£à¸²à¸Ÿà¸ªà¸£à¸¸à¸› à¹à¸•à¹ˆà¸¢à¸±à¸‡à¸„à¸‡à¹„à¸§à¹‰à¸•à¸²à¸¡à¹‚à¸ˆà¸—à¸¢à¹Œ)
df_sorted = df.sort_values(by='customer_id' , ascending=True)


# 4. à¸‚à¸±à¹‰à¸™à¸•à¸­à¸™à¸à¸²à¸£à¸ªà¸£à¸¸à¸›à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸žà¸·à¹ˆà¸­à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿ (à¸à¸²à¸£à¸šà¹‰à¸²à¸™)
# à¸ªà¸£à¸¸à¸›: à¸ˆà¸±à¸”à¸à¸¥à¸¸à¹ˆà¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸•à¸²à¸¡ 'product' à¹à¸¥à¸°à¸«à¸²à¸œà¸¥à¸£à¸§à¸¡à¸‚à¸­à¸‡ 'revenue'
df_summary = df_sorted.groupby('product')['revenue'].sum().sort_values(ascending=False)

# 5. à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿà¹à¸—à¹ˆà¸‡ (Matplotlib Plotting)
plt.figure(figsize=(10, 6)) # à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸à¸£à¸²à¸Ÿ

# à¹ƒà¸Šà¹‰ Pandas plot wrapper à¸—à¸µà¹ˆà¸—à¸³à¸‡à¸²à¸™à¸£à¹ˆà¸§à¸¡à¸à¸±à¸š Matplotlib
df_summary.plot(kind='bar', color='skyblue')

# à¸à¸³à¸«à¸™à¸”à¸£à¸²à¸¢à¸¥à¸°à¹€à¸­à¸µà¸¢à¸”à¸à¸£à¸²à¸Ÿ
plt.title('Total Revenue by Product Category (Matplotlib Summary)')
plt.xlabel('Product')
plt.ylabel('Total Revenue (Million THB/USD)')
plt.xticks(rotation=45, ha='right') # à¸«à¸¡à¸¸à¸™à¸‰à¸¥à¸²à¸à¹à¸à¸™ X à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¸­à¹ˆà¸²à¸™à¸‡à¹ˆà¸²à¸¢
plt.grid(axis='y', linestyle='--', alpha=0.7) # à¹€à¸žà¸´à¹ˆà¸¡à¹€à¸ªà¹‰à¸™à¸à¸£à¸´à¸”
plt.tight_layout() # à¸›à¸£à¸±à¸šà¹ƒà¸«à¹‰à¸à¸£à¸²à¸Ÿà¹à¸ªà¸”à¸‡à¸œà¸¥à¹„à¸”à¹‰à¸žà¸­à¸”à¸µ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ... à¹‚à¸„à¹‰à¸”à¸ªà¸£à¹‰à¸²à¸‡ df_sorted à¹à¸¥à¸° df_summary à¹€à¸”à¸´à¸¡ ...

# 1. à¸ªà¸£à¸¸à¸›à¸‚à¹‰à¸­à¸¡à¸¹à¸¥: à¸ˆà¸±à¸”à¸à¸¥à¸¸à¹ˆà¸¡à¸•à¸²à¸¡ 'product' à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹à¸à¸™ X à¹€à¸›à¹‡à¸™à¸Šà¸·à¹ˆà¸­à¸ªà¸´à¸™à¸„à¹‰à¸²
df_summary = df_sorted.groupby('product')['revenue'].sum().sort_values(ascending=False)

# 2. à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿà¹€à¸ªà¹‰à¸™ (Line Plot) à¸ˆà¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸ªà¸£à¸¸à¸›
plt.figure(figsize=(10, 6))

# à¹ƒà¸Šà¹‰ .plot(kind='line') à¸šà¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸ªà¸£à¸¸à¸› (df_summary)
df_summary.plot(kind='line', marker='o', linestyle='-', color='purple') 
# 'marker' à¹à¸¥à¸° 'linestyle' à¸–à¸¹à¸à¹€à¸žà¸´à¹ˆà¸¡à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹€à¸«à¹‡à¸™à¸ˆà¸¸à¸”à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸Šà¸±à¸”à¹€à¸ˆà¸™à¸‚à¸¶à¹‰à¸™

# à¸à¸³à¸«à¸™à¸”à¸£à¸²à¸¢à¸¥à¸°à¹€à¸­à¸µà¸¢à¸”à¸à¸£à¸²à¸Ÿ
plt.title('Total Revenue by Product Category (Line Plot)')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right') # à¸«à¸¡à¸¸à¸™à¸Šà¸·à¹ˆà¸­à¸ªà¸´à¸™à¸„à¹‰à¸²à¸šà¸™à¹à¸à¸™ X
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¹‚à¸«à¸¥à¸”à¹à¸¥à¸°à¹€à¸•à¸£à¸µà¸¢à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
# *** à¹‚à¸›à¸£à¸”à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸š Path à¸‚à¸­à¸‡à¹„à¸Ÿà¸¥à¹Œ '../data_raw/sample_sales.csv' à¹ƒà¸«à¹‰à¸–à¸¹à¸à¸•à¹‰à¸­à¸‡à¸à¹ˆà¸­à¸™à¸£à¸±à¸™ ***
df = pd.read_csv('../data_raw/sample_sales.csv')

# 2. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¸„à¸³à¸™à¸§à¸“à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'revenue'
df['revenue'] = df['quantity'] * df['unit_price'] * (1 - df['discount'])

# 3. à¹‚à¸„à¹‰à¸”à¸ªà¸³à¸«à¸£à¸±à¸šà¹€à¸£à¸µà¸¢à¸‡à¸¥à¸³à¸”à¸±à¸š (df_sorted)
df_sorted = df.sort_values(by='customer_id' , ascending=True)


# 4. à¸ªà¸£à¸¸à¸›à¸‚à¹‰à¸­à¸¡à¸¹à¸¥: à¸ˆà¸±à¸”à¸à¸¥à¸¸à¹ˆà¸¡à¸•à¸²à¸¡ 'product' à¹€à¸žà¸·à¹ˆà¸­à¸«à¸²à¸œà¸¥à¸£à¸§à¸¡à¸‚à¸­à¸‡à¸£à¸²à¸¢à¹„à¸”à¹‰
df_summary = df_sorted.groupby('product')['revenue'].sum().sort_values(ascending=False)

# 5. à¸ªà¸£à¹‰à¸²à¸‡à¸à¸£à¸²à¸Ÿà¸§à¸‡à¸à¸¥à¸¡ (Pie Chart)
plt.figure(figsize=(8, 8)) # à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸à¸£à¸²à¸Ÿà¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™à¸ªà¸µà¹ˆà¹€à¸«à¸¥à¸µà¹ˆà¸¢à¸¡à¸ˆà¸±à¸•à¸¸à¸£à¸±à¸ª

# à¹ƒà¸Šà¹‰ .plot(kind='pie') à¹€à¸žà¸·à¹ˆà¸­à¸žà¸¥à¸­à¸•à¸à¸£à¸²à¸Ÿà¸§à¸‡à¸à¸¥à¸¡
df_summary.plot(
    kind='pie',
    autopct='%1.1f%%', # à¹à¸ªà¸”à¸‡à¸ªà¸±à¸”à¸ªà¹ˆà¸§à¸™à¹€à¸›à¹‡à¸™à¹€à¸›à¸­à¸£à¹Œà¹€à¸‹à¹‡à¸™à¸•à¹Œ (à¸—à¸¨à¸™à¸´à¸¢à¸¡ 1 à¸•à¸³à¹à¸«à¸™à¹ˆà¸‡)
    startangle=90,     # à¹€à¸£à¸´à¹ˆà¸¡à¸•à¹‰à¸™à¸à¸²à¸£à¸žà¸¥à¸­à¸•à¸—à¸µà¹ˆ 12 à¸™à¸²à¸¬à¸´à¸à¸²
    wedgeprops={'edgecolor': 'black'} # à¹€à¸žà¸´à¹ˆà¸¡à¹€à¸ªà¹‰à¸™à¸‚à¸­à¸šà¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸Šà¸´à¹‰à¸™
)

# à¸à¸³à¸«à¸™à¸”à¸£à¸²à¸¢à¸¥à¸°à¹€à¸­à¸µà¸¢à¸”à¸à¸£à¸²à¸Ÿ
plt.title('Proportion of Total Revenue by Product Category')
plt.ylabel('') # à¸¥à¸šà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸Šà¸·à¹ˆà¸­ 'revenue' à¸­à¸­à¸à¸ˆà¸²à¸à¹à¸à¸™ Y
plt.show()
```
